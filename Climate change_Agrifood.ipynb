### Time Series Analysis, VAR model

with two variables, Agrifood Systems Waste Disposal and IPPU

```python
import statsmodels.api as sm
from statsmodels.tsa.api import VAR
from statsmodels.tsa.stattools import adfuller
```

```python
df = agrifood[['Agrifood Systems Waste Disposal', 'IPPU', 'TemperatureDeviation']]
df.index = agrifood["Year"]
df.head()
```

```python
df.describe()
```

```python
df.plot(figsize = (8,5))
```

```python
df['TemperatureDeviation'].plot()
```

```python
# stationarity
adf_ASWD = adfuller(df['Agrifood Systems Waste Disposal'])
adf_IPPU = adfuller(df['IPPU'])
adf_TD = adfuller(df['TemperatureDeviation'])
print("p-value for Agrifood Systems Waste Disposal: {}".format(adf_ASWD[1]))
print("p-value for IPPU: {}".format(adf_IPPU[1]))
print("p-value for TemperatureDeviation: {}".format(adf_TD[1]))
```

```python
# differencing the data (x(t)-x(t-1)) until we get a point where the series is stationary
df_stationary = df.diff().dropna()
```

```python
df_stationary.plot()
```

```python
plt.plot(df_stationary['TemperatureDeviation'])
```

```python
# after differentiation
adf_ASWD_diff = adfuller(df_stationary['Agrifood Systems Waste Disposal'])
adf_IPPU_diff = adfuller(df_stationary['IPPU'])
adf_TD_diff = adfuller(df_stationary['TemperatureDeviation'])
print("p-value for Agrifood Systems Waste Disposal: {}".format(adf_ASWD_diff[1]))
print("p-value for IPPU: {}".format(adf_IPPU_diff[1]))
print("p-value for TemperatureDeviation: {}".format(adf_TD_diff[1]))
```

```python
train = df_stationary.iloc[:18]
test = df_stationary.iloc[18:]
```

```python
train.shape
```

```python
test.shape
```

```python
forecasting_model = VAR(train)
```

```python
# optimal lag order
forecasting_model = VAR(train)
x = forecasting_model.select_order(maxlags=3)
x.summary()
```

```python
results = forecasting_model.fit(3)
results.summary()
```

```python
laaged_values = train.values[-3:]
forecast = pd.DataFrame(results.forecast(y= laaged_values, steps=8), index = test.index, columns= ['Agrifood Systems Waste Disposal_d', 'IPPU_d', 'TemperatureDeviation_d'])
forecast
```

```python
forecast["Agrifood Systems Waste Disposal_f"] = df["Agrifood Systems Waste Disposal"].iloc[-8-1] + forecast['Agrifood Systems Waste Disposal_d'].cumsum()
forecast["IPPU_f"] = df["IPPU"].iloc[-8-1] + forecast['IPPU_d'].cumsum()
forecast["TemperatureDeviation_f"] = df["TemperatureDeviation"].iloc[-8-1] + forecast['TemperatureDeviation_d'].cumsum()
forecast
```

```python
test = df.iloc[-8:,:]
test["Agrifood Systems Waste Disposal_f"] = forecast["Agrifood Systems Waste Disposal_f"]
test["IPPU_f"] = forecast["IPPU_f"]
test["TemperatureDeviation_f"] = forecast["TemperatureDeviation_f"]
test.plot()
```

```python
test = df.iloc[-8:,:1]
test["Agrifood Systems Waste Disposal_f"] = forecast["Agrifood Systems Waste Disposal_f"]
test.plot()
```

```python
test = df.iloc[:, :1]
test["Agrifood Systems Waste Disposal_f"] = forecast["Agrifood Systems Waste Disposal_f"]
test.plot()
plt.ylabel("CO2 Emission")
plt.legend(['Agrifood Systems Waste Disposal', 'forecasted'])
```

```python
results = forecasting_model.fit(3)
results.summary()
```

```python
laaged_values = train.values[-3:]
forecast2 = pd.DataFrame(results.forecast(y= laaged_values, steps=10), columns= ['Agrifood Systems Waste Disposal_d', 'IPPU_d', 'TemperatureDeviation_d'])
forecast2
```

```python
forecast2["Agrifood Systems Waste Disposal_f"] = df["Agrifood Systems Waste Disposal"].iloc[-1] + forecast2['Agrifood Systems Waste Disposal_d'].cumsum()
forecast2["IPPU_f"] = df["IPPU"].iloc[-1] + forecast2['IPPU_d'].cumsum()
forecast2["TemperatureDeviation_f"] = df["TemperatureDeviation"].iloc[-1] + forecast2['TemperatureDeviation_d'].cumsum()
forecast2
```

```python
temp = df.iloc[:, 2:3]
pred = pd.DataFrame(forecast2['TemperatureDeviation_f'])
pred = pred.set_index([pd.Index([2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024, 2025, 2026])])
pred.rename(columns={"TemperatureDeviation_f": "TemperatureDeviation"}, inplace=True)
pred = pd.concat([temp, pred])
pred
```

```python
pred.plot()
plt.xlabel("Year")
plt.ylabel("TemperatureDeviation")
plt.legend('', frameon=False)
```
