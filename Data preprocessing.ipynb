### Agri-food CO2 Emission Data

```python
agrifood = pd.read_csv('Agrifood_co2_emission.csv')
agrifood
```

```python
agrifood.isnull().sum()
```

__missing values__

```python
agrifood[agrifood['Savanna fires'].isnull()]
```

```python
# Savanna fires null values: Holy See
# It is not possible that fire events, manure, and farm occur in the city of Vatican
null = agrifood[agrifood['Area']=='Holy See'].isnull()
null.sum()
```

```python
# fill with 0
agrifood.loc[agrifood['Area'] == 'Holy See'] = agrifood[agrifood['Area'] == 'Holy See'].fillna(0)
```

```python
agrifood[agrifood['Forest fires'].isnull()]
```

```python
# It is not that fire events did not occur in Monaco and San Marino
# fill with 0
agrifood.loc[(agrifood['Area'] == 'Monaco')|(agrifood['Area'] == 'San Marino'), ['Forest fires']] = agrifood.loc[(agrifood['Area'] == 'Monaco') | (agrifood['Area']=='San Marino'), ['Forest fires']].fillna(0)
```

```python
agrifood[agrifood['Fires in humid tropical forests'].isnull()]
```

```python
# It is possible that fire events in tropical forests did not occur in Channel Islands, Liechtenstein, Monaco, San Marino
# fill with 0
agrifood.loc[(agrifood['Area'] == 'Channel Islands')|(agrifood['Area'] == 'Liechtenstein')|(agrifood['Area'] == 'Monaco')|(agrifood['Area'] == 'San Marino'), ['Fires in humid tropical forests']] = agrifood.loc[(agrifood['Area'] == 'Channel Islands')|(agrifood['Area'] == 'Liechtenstein')|(agrifood['Area'] == 'Monaco')|(agrifood['Area'] == 'San Marino'), ['Fires in humid tropical forests']].fillna(0)
```

```python
# fill all missing values with 0
# MNAR (Missing Data Not at Random)
agrifood = agrifood.fillna(0)
```

```python
agrifood.isnull().sum()
```

__groupby__

```python
agrifood = agrifood.groupby('Year').sum()
agrifood['Year'] = agrifood.index
```

__removing variables__

```python
# 'Crop Residues', 'Manure applied to Soils', Manure Management', and 'On-farm energy use' : lots of missing values
# 'Forestland', 'Net Forest conversion', 'Rural population', 'Urban population', 'Total Population - Male',
# 'Total Population - Female', 'total_emission', 'Average Temperature' : not related to CO2 emission
df_agrifood = agrifood[['Year', 'Savanna fires', 'Forest fires', 'Rice Cultivation', 'Drained organic soils (CO2)',
                             'Pesticides Manufacturing', 'Food Transport', 'Food Household Consumption', 'Food Retail',
                             'On-farm Electricity Use', 'Food Packaging', 'Agrifood Systems Waste Disposal', 'Food Processing',
                             'Fertilizers Manufacturing', 'IPPU', 'Manure left on Pasture',
                             'Fires in organic soils', 'Fires in humid tropical forests']]
df_agrifood = df_agrifood.reset_index(drop=True)
```

```python
df_agrifood
```

__with temperature__

```python
df_agrifood2 = pd.merge(df_agrifood, temperature, on="Year", how="inner")
df_agrifood2
```

```python
df_agrifood2.to_csv('agrifood.csv', index=False)
```

__Standardisation__

```python
pre = df_agrifood2.iloc[:, 1:-1]
year = df_agrifood2.iloc[:, 0]
temp = df_agrifood2.iloc[:, -1]
```

```python
from sklearn import preprocessing

scaler = preprocessing.StandardScaler().fit(pre)
```

```python
scaled = scaler.transform(pre)
scaled = pd.DataFrame(scaled)
scaled.rename(columns={0: 'Savanna fires', 1: 'Forest fires', 2: 'Rice Cultivation',
       3: 'Drained organic soils (CO2)', 4: 'Pesticides Manufacturing',
       5: 'Food Transport', 6: 'Food Household Consumption', 7: 'Food Retail',
       8: 'On-farm Electricity Use', 9: 'Food Packaging',
       10: 'Agrifood Systems Waste Disposal', 11: 'Food Processing',
       12: 'Fertilizers Manufacturing', 13: 'IPPU',
       14: 'Manure left on Pasture', 15: 'Fires in organic soils',
       16: 'Fires in humid tropical forests'}, inplace=True)
scaled
```

```python
agrifood_scaled = pd.concat([year, scaled], axis=1)
agrifood_scaled = pd.concat([agrifood_scaled, temp], axis=1)
agrifood_scaled
```

```python
agrifood_scaled.to_csv('agrifood_scaled.csv', index=False)
```
